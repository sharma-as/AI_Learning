```python
from transformers import pipeline
import random

# Base model (simulating an "actor")
generator = pipeline("text-generation", model="gpt2")

# Memory to store reflections
reflections = []

def reflexion_agent(task, n_iters=3):
    global reflections
    print(f"Task: {task}")

    for i in range(n_iters):
        print(f"\n--- Iteration {i+1} ---")

        # Include reflections in the prompt
        context = " ".join(reflections)
        prompt = f"{context}\nSolve this task: {task}"

        output = generator(prompt, max_length=80, num_return_sequences=1)[0]['generated_text']
        print("Agent output:", output)

        # Simple evaluation: random success/failure
        success = random.choice([True, False])

        if success:
            print("✅ Success!")
            break
        else:
            critique = f"Reflection: Last attempt failed because reasoning was unclear."
            reflections.append(critique)
            print("❌ Failure. Adding reflection.")

    return output

# Example
reflexion_agent("Translate 'bonjour' to English")
```
